```python
def compare(arr1, arr2):
  if arr1 < arr2:
    return 1
  else:
    return 0

compare(2, 3)
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([2, 3, 0, 1, 4])

compare(arr1, arr2)
```

```python
#vectorize
vectorize_compare = np.vectorize(compare)
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([2, 3, 0, 1, 4])
vectorize_compare(arr1, arr2)
```

```python
#circle area
def caculate_circle_area(radius):
  return 3.14*radius**2

```

```python
caculate_circle_area(3)
```

```python
vectorize_caculate_circle_area = np.vectorize(caculate_circle_area)
```

```python
radius_arr = np.array([1, 2, 3, 4])
vectorize_caculate_circle_area(radius_arr)
```
